```python
a=int(input("enter number A : "))
b=int(input("enter number B : "))
symbol=input("enter a symbol + - * % / : ")
#carry out operation based on symbol provided if none of the given symbol provided print invalid symbol
if symbol=="+":
  print(f"{a}+{b}={a+b}")
elif symbol=="-":
  print(f"{a}-{b}={a-b}")
elif symbol=="*":
  print(f"{a}*{b}={a*b}")
elif symbol=="%":
  print(f"{a}%{b}={a%b}")
elif symbol=="/":
  print(f"{a}/{b}={a/b}")
else:
  print("invalid symbol")

```

```python
# Initialize an empty list
numbers = []

inp = input("Enter a number (or press Enter to finish): ")

while inp != "":
    numbers.append(int(inp))
    inp = input("Enter a number (or press Enter to finish): ")

# Print the list of numbers
print(numbers)

# Print the square of each number in the list
for i in numbers:
    print(f"{i}² = {i**2}")
```

```python
n=int(input("enter a number : "))
fib_series = [0, 1]  # Starting values for the Fibonacci sequence
for i in range(2, n): # Generate the Fibonacci sequence
  next_num = fib_series[-1] + fib_series[-2] # Calculate the next number by adding the last two numbers
  fib_series.append(next_num) #add it to the list

print(fib_series) #print the fibonacci series
```
